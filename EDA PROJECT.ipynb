#### Handling missing values

```python
df.isnull().sum()
```

```python
df.torque.isnull().sum()
```

```python
df.torque.value_counts(normalize=True)
```

```python
torque_mode=df.torque.mode()[0]
torque_mode
```

```python
df.torque.fillna(torque_mode,inplace=True)
df.torque.value_counts(normalize=True)
```

#### I have done mode value for null containing column(torque) and filled the missing values with   the mode values. value counts give the count of torque. 

#  

```python
df.torque.isnull().sum()
```

#### After filling the null values in torque column it became zero which means there are no missing values present in the torque column.in that way I have filled all the null values in null containing column in this way.

```python
mileage_mode=df.mileage.mode()[0]
mileage_mode
```

```python
df.mileage.fillna(mileage_mode,inplace=True)
df.mileage.value_counts(normalize=True)
```

```python
df.mileage.isnull().sum()
```

```python
engine_mode=df.engine.mode()[0]
engine_mode
```

```python
df.engine.fillna(engine_mode,inplace=True)
df.engine.value_counts(normalize=True)
```

```python
df.engine.isnull().sum()
```

```python
seats_mode=df.seats.mode()[0]
seats_mode
```

```python
df.seats.fillna(seats_mode,inplace=True)
df.seats.value_counts(normalize=True)
```

```python
df.seats.isnull().sum()
```

```python
df.isnull().sum()
```
